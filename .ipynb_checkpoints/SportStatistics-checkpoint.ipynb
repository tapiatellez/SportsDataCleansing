## Question 1
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NHL** using **2018** data.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

nhl_df=pd.read_csv("assets/nhl.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def nhl_correlation(): 
    # YOUR CODE HERE
    # Working with nhl data
    # Only data from 2018
    nhl_df=pd.read_csv("assets/nhl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    nhl_df = nhl_df[nhl_df['year'] == 2018]
    nhl_df['team name'] = nhl_df['team'].str.split().str[-1]
    nhl_df['area'] = nhl_df.team.str.rsplit(' ', 1).str[0]
    #Remove asterix
    nhl_df['team name'] = nhl_df['team name'].str.replace('*', '')
    # Replace values by hand for nhl_df
    nhl_df['team name'] = nhl_df['team name'].replace(['Leafs', 'Wings', 'Jackets', 'Knights'],
                                                      ['Maple Leafs', 'Red Wings', 'Blue Jackets', 'Golden Knights'])
    nhl_df['area'] = nhl_df['area'].replace(['Toronto Maple', 'Detroid Red', 'Columbus Blue', 'Vegas Golden', 'San Jose', 'New Jersey', 'Anaheim'],
                                            ['Toronto', 'Detroit', 'Columbus', 'Las Vegas', 'San Francisco', 'New York', 'Los Angeles'])
    nhl_df_wl = nhl_df[['team name', 'area', 'W', 'L']] # work with important columns
    nhl_df_wl = nhl_df_wl.set_index('team name') # set index to the team name
    #print(nhl_df_wl)
    # working with cities data
    cities_nhl = cities.drop(['NFL', 'MLB', 'NBA'], axis = 1)
    cities_nhl.rename(columns={'Metropolitan area': 'M. area', 'Population (2016 est.)[8]': 'Population'}, inplace = True)
    cities_nhl['team name'] = cities_nhl['NHL'].str.replace(r"\[(.*?)\]", "")
    cities_nhl['M. area'] = cities_nhl['M. area'].replace(["New York City", 'San Francisco Bay Area', 'Dallas–Fort Worth', 'Washington, D.C.', 'Minneapolis–Saint Paul', 'Miami–Fort Lauderdale', 'Tampa Bay Area'],
                                                          ['New York', 'San Francisco', 'Dallas', 'Washington', 'Minneapolis', 'Miami', 'Tampa Bay'])
    cities_nhl['team name'] = cities_nhl['team name'].astype(str)
    cities_nhl = cities_nhl.set_index(['M. area', 'Population', 'NHL'])
    cities_nhl = cities_nhl.apply(lambda x: x.str.split('[a-z][A-Z]').explode()).reset_index()
    cities_nhl['team name'] = cities_nhl['team name'].replace(['Ranger','slander', 'evils', 'King', 'ucks'],
                                                          ['Rangers', 'Islanders', 'Devils', 'Kings', 'Ducks'])
    cities_nhl = cities_nhl.set_index('team name')
    #print(cities_nhl)
    # Merging
    merged_df = pd.merge(cities_nhl, nhl_df_wl, how = 'inner', left_index = True, right_index = True)
    merged_df = merged_df.reset_index()
    merged_df = merged_df.sort_values('M. area')
    merged_df = merged_df.set_index('M. area')
    merged_df['W'] = pd.to_numeric(merged_df['W'], downcast = 'float')
    merged_df['L'] = pd.to_numeric(merged_df['L'], downcast = 'float')
    merged_df['Population'] = pd.to_numeric(merged_df['Population'], downcast = 'float')
    merged_df['WL Ratio'] = merged_df['W']/(merged_df['W'] + merged_df['L'])
    merged_df = merged_df.drop(['team name', 'NHL', 'area', 'W', 'L'], axis = 1)
    merged_df = merged_df.groupby(['M. area', 'Population'])[['WL Ratio']].mean()
    merged_df = merged_df.reset_index()
    #print(merged_df)
    
    population_by_region = merged_df['Population'] # pass in metropolitan area population from cities
    win_loss_by_region = merged_df['WL Ratio'] # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
```

## Question 2
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NBA** using **2018** data.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

nba_df=pd.read_csv("assets/nba.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def nba_correlation():
    # YOUR CODE HERE
    # Archives
    nba_df=pd.read_csv("assets/nba.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    # NBA Data
    # Only data from 2018
    nba_df = nba_df[nba_df['year'] == 2018]
    nba_df['team'] = nba_df['team'].str.replace(r'\((.*?)\)', '')
    nba_df['team'] = nba_df['team'].str.replace('*', '')
    nba_df['team name'] = nba_df['team'].str.split().str[-1]
    nba_df['team name'] = nba_df['team name'].str.replace('*', '')
    nba_df['area'] = nba_df.team.str.rsplit(' ', 1).str[0]
    nba_df = nba_df[['team', 'W', 'L', 'team name', 'area']]
    nba_df['team name'] = nba_df['team name'].replace(['Warriors'],
                                     ['Golden State Warriors'])
    nba_df['area'] = nba_df['area'].replace(['Golden State', 'Portland Trail'],
                                ['San Francisco', 'Portland'])
    nba_df = nba_df.set_index('team name')
    # Cities Data
    cities_nba = cities.drop(['NHL', 'MLB', 'NFL'], axis = 1)
    cities_nba.rename(columns={'Metropolitan area': 'M. area', 'Population (2016 est.)[8]': 'Population', 'NBA': 'team'}, inplace = True)
    cities_nba['team'] = cities_nba['team'].str.replace(r'\[(.*?)\]', '')
    cities_nba = (cities_nba.set_index(['M. area', 'Population']).apply(lambda x: x.str.split('[a-z][A-Z]').explode()).reset_index())
    cities_nba['team'] = cities_nba['team'].replace(['Knick', 'ets', 'Laker', 'lippers', 'Warriors', 'Trail Blazers'],
                                                ['Knicks', 'Nets', 'Lakers', 'Clippers', 'Golden State Warriors', 'Blazers'])
    cities_nba = cities_nba.set_index('team')
    # Merging, Ratio and Correlation
    merged_df = pd.merge(cities_nba, nba_df, how = 'inner', left_index = True, right_index = True)
    merged_df = merged_df.reset_index()
    merged_df = merged_df.sort_values('M. area')
    merged_df = merged_df.set_index('M. area')
    merged_df['W'] = pd.to_numeric(merged_df['W'], downcast = 'float')
    merged_df['L'] = pd.to_numeric(merged_df['L'], downcast = 'float')
    merged_df['Population'] = pd.to_numeric(merged_df['Population'], downcast = 'float')
    merged_df['WL Ratio'] = merged_df['W']/(merged_df['W'] + merged_df['L'])
    merged_df = merged_df.drop(['team', 'index', 'area', 'W', 'L'], axis = 1)
    merged_df = merged_df.groupby(['M. area', 'Population'])[['WL Ratio']].mean()
    merged_df = merged_df.reset_index()
    # Correlation
    population_by_region = merged_df['Population'] # pass in metropolitan area population from cities
    win_loss_by_region = merged_df['WL Ratio'] # pass in win/loss ratio from nba_df in the same order as cities["Metropolitan area"]

    assert len(population_by_region) == len(win_loss_by_region), "Q2: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q2: There should be 28 teams being analysed for NBA"

    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
```

## Question 3

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

mlb_df=pd.read_csv("assets/mlb.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def mlb_correlation(): 
    # YOUR CODE HERE
    # MLB Data
    mlb = mlb_df[mlb_df['year'] == 2018]
    mlb['team name'] = mlb['team'].str.split().str[-1]
    mlb['area'] = mlb.team.str.rsplit(' ', 1).str[0]
    mlb = mlb[['team', 'W', 'L', 'team name', 'area']]
    mlb['team name'] = mlb['team name'].replace(['Sox', 'Rays', 'Jays', ''],
                                                ['Red Sox', 'Bay Rays', 'Blue Jays', ''])
    mlb['team name'] = mlb['team name'].replace([])
    mlb['team name'][8] = 'White Sox'
    mlb = mlb.set_index('team name')
    # Cities Data
    cities_mlb = cities.drop(['NHL', 'NBA', 'NFL'], axis = 1)
    cities_mlb.rename(columns={'Metropolitan area': 'M. area', 'Population (2016 est.)[8]': 'Population', 'MLB': 'team name'}, inplace = True)
    cities_mlb['team name'] = cities_mlb['team name'].str.replace(r'\[(.*?)\]', '')
    cities_mlb = (cities_mlb.set_index(['M. area', 'Population']).apply(lambda x: x.str.split('[a-z][A-Z]').explode()).reset_index())
    cities_mlb['team name'] = cities_mlb['team name'].replace(['Yankee', 'ets', 'Dodger', 'ngels', 'Giant', 'thletics', 'Cub', 'hite Sox', 'Rays'],
                                                ['Yankees', 'Mets', 'Dodgers', 'Angels', 'Giants', 'Athletics', 'Cubs', 'White Sox', 'Bay Rays'])
    cities_mlb = cities_mlb.set_index('team name')
    # Merging and WL-Ratio
    merged_df = pd.merge(cities_mlb, mlb, how = 'inner', left_index = True, right_index = True)
    merged_df = merged_df.reset_index()
    merged_df = merged_df.sort_values('M. area')
    merged_df = merged_df.set_index('M. area')
    merged_df['W'] = pd.to_numeric(merged_df['W'], downcast = 'float')
    merged_df['L'] = pd.to_numeric(merged_df['L'], downcast = 'float')
    merged_df['Population'] = pd.to_numeric(merged_df['Population'], downcast = 'float')
    merged_df['WL Ratio'] = merged_df['W']/(merged_df['W'] + merged_df['L'])
    merged_df
    merged_df = merged_df.drop(['team name', 'team', 'area', 'W', 'L'], axis = 1)
    merged_df = merged_df.groupby(['M. area', 'Population'])[['WL Ratio']].mean()
    merged_df = merged_df.reset_index()
    merged_df
    #Correlation
    population_by_region = merged_df['Population'] # pass in metropolitan area population from cities
    win_loss_by_region = merged_df['WL Ratio'] # pass in win/loss ratio from mlb_df in the same order as cities["Metropolitan area"]

    assert len(population_by_region) == len(win_loss_by_region), "Q3: Your lists must be the same length"
    assert len(population_by_region) == 26, "Q3: There should be 26 teams being analysed for MLB"

    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

mlb_df=pd.read_csv("assets/mlb.csv")
nhl_df=pd.read_csv("assets/nhl.csv")
nba_df=pd.read_csv("assets/nba.csv")
nfl_df=pd.read_csv("assets/nfl.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def sports_team_performance():
    # YOUR CODE HERE
    #Import all the data
    mlb_df=pd.read_csv("assets/mlb.csv")
    nhl_df=pd.read_csv("assets/nhl.csv")
    nba_df=pd.read_csv("assets/nba.csv")
    nfl_df=pd.read_csv("assets/nfl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    # NBA data
    # Only data from 2018
    nba_df = nba_df[nba_df['year'] == 2018]
    nba_df['team'] = nba_df['team'].str.replace(r'\((.*?)\)', '')
    nba_df['team'] = nba_df['team'].str.replace('*', '')
    nba_df['team name'] = nba_df['team'].str.split().str[-1]
    nba_df['team name'] = nba_df['team name'].str.replace('*', '')
    nba_df['area'] = nba_df.team.str.rsplit(' ', 1).str[0]
    nba_df = nba_df[['team', 'W', 'L', 'team name', 'area']]
    nba_df['team name'] = nba_df['team name'].replace(['Warriors'],
                                     ['Golden State Warriors'])
    #nba_df['area'] = nba_df['area'].replace(['Golden State', 'Portland Trail'],
    #                            ['San Francisco', 'Portland'])
    nba_df = nba_df.set_index('team name')
    # Cities Data
    cities_nba = cities.drop(['NHL', 'MLB', 'NFL'], axis = 1)
    cities_nba.rename(columns={'Metropolitan area': 'M. area', 'Population (2016 est.)[8]': 'Population', 'NBA': 'team'}, inplace = True)
    cities_nba['team'] = cities_nba['team'].str.replace(r'\[(.*?)\]', '')
    cities_nba = (cities_nba.set_index(['M. area', 'Population']).apply(lambda x: x.str.split('[a-z][A-Z]').explode()).reset_index())
    cities_nba['team'] = cities_nba['team'].replace(['Knick', 'ets', 'Laker', 'lippers', 'Warriors', 'Trail Blazers'],
                                                ['Knicks', 'Nets', 'Lakers', 'Clippers', 'Golden State Warriors', 'Blazers'])
    cities_nba = cities_nba.set_index('team')
    # Merging, Ratio and Correlation
    merged_nba = pd.merge(cities_nba, nba_df, how = 'inner', left_index = True, right_index = True)
    merged_nba = merged_nba.reset_index()
    merged_nba = merged_nba.sort_values('M. area')
    merged_nba = merged_nba.set_index('M. area')
    merged_nba['W'] = pd.to_numeric(merged_nba['W'], downcast = 'float')
    merged_nba['L'] = pd.to_numeric(merged_nba['L'], downcast = 'float')
    merged_nba['Population'] = pd.to_numeric(merged_nba['Population'], downcast = 'float')
    merged_nba['WL Ratio'] = merged_nba['W']/(merged_nba['W'] + merged_nba['L'])
    merged_nba = merged_nba.drop(['team', 'index', 'area', 'W', 'L'], axis = 1)
    merged_nba = merged_nba.groupby(['M. area', 'Population'])[['WL Ratio']].mean()
    merged_nba = merged_nba.reset_index()
    #merged_nba['M. area'] = merged_nba['M. area'].replace(['Dallas–Fort Worth', 'Miami–Fort Lauderdale', 'Minneapolis–Saint Paul', 'New York City', 'Oklahoma City', 'Salt Lake City', 'San Francisco Bay Area', 'Washington, D.C.'],
    #                                                      ['Dallas', 'Miami', 'Minneapolis', 'New York', 'Oklahoma', 'Salt Lake', 'San Francisco', 'Washington'])
    merged_nba = merged_nba.set_index('M. area')
    merged_nba = merged_nba.rename(columns = {'WL Ratio': 'WL Ratio NBA'})
    
    #NHL Data 
    # Working with nhl data
    # Only data from 2018
    nhl_df=pd.read_csv("assets/nhl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    nhl_df = nhl_df[nhl_df['year'] == 2018]
    nhl_df['team name'] = nhl_df['team'].str.split().str[-1]
    nhl_df['area'] = nhl_df.team.str.rsplit(' ', 1).str[0]
    #Remove asterix
    nhl_df['team name'] = nhl_df['team name'].str.replace('*', '')
    # Replace values by hand for nhl_df
    nhl_df['team name'] = nhl_df['team name'].replace(['Leafs', 'Wings', 'Jackets', 'Knights'],
                                                      ['Maple Leafs', 'Red Wings', 'Blue Jackets', 'Golden Knights'])
    nhl_df['area'] = nhl_df['area'].replace(['Toronto Maple', 'Detroid Red', 'Columbus Blue', 'Vegas Golden', 'San Jose', 'New Jersey', 'Anaheim'],
                                            ['Toronto', 'Detroit', 'Columbus', 'Las Vegas', 'San Francisco', 'New York', 'Los Angeles'])
    nhl_df_wl = nhl_df[['team name', 'area', 'W', 'L']] # work with important columns
    nhl_df_wl = nhl_df_wl.set_index('team name') # set index to the team name
    #print(nhl_df_wl)
    # working with cities data
    cities_nhl = cities.drop(['NFL', 'MLB', 'NBA'], axis = 1)
    cities_nhl.rename(columns={'Metropolitan area': 'M. area', 'Population (2016 est.)[8]': 'Population'}, inplace = True)
    cities_nhl['team name'] = cities_nhl['NHL'].str.replace(r"\[(.*?)\]", "")
    #cities_nhl['M. area'] = cities_nhl['M. area'].replace(["New York City", 'San Francisco Bay Area', 'Dallas–Fort Worth', 'Washington, D.C.', 'Minneapolis–Saint Paul', 'Miami–Fort Lauderdale', 'Tampa Bay Area'],
    #                                                      ['New York', 'San Francisco', 'Dallas', 'Washington', 'Minneapolis', 'Miami', 'Tampa Bay'])
    cities_nhl['team name'] = cities_nhl['team name'].astype(str)
    cities_nhl = cities_nhl.set_index(['M. area', 'Population', 'NHL'])
    cities_nhl = cities_nhl.apply(lambda x: x.str.split('[a-z][A-Z]').explode()).reset_index()
    cities_nhl['team name'] = cities_nhl['team name'].replace(['Ranger','slander', 'evils', 'King', 'ucks'],
                                                          ['Rangers', 'Islanders', 'Devils', 'Kings', 'Ducks'])
    cities_nhl = cities_nhl.set_index('team name')
    #print(cities_nhl)
    # Merging
    merged_nhl = pd.merge(cities_nhl, nhl_df_wl, how = 'inner', left_index = True, right_index = True)
    merged_nhl = merged_nhl.reset_index()
    merged_nhl = merged_nhl.sort_values('M. area')
    merged_nhl = merged_nhl.set_index('M. area')
    merged_nhl['W'] = pd.to_numeric(merged_nhl['W'], downcast = 'float')
    merged_nhl['L'] = pd.to_numeric(merged_nhl['L'], downcast = 'float')
    merged_nhl['Population'] = pd.to_numeric(merged_nhl['Population'], downcast = 'float')
    merged_nhl['WL Ratio'] = merged_nhl['W']/(merged_nhl['W'] + merged_nhl['L'])
    merged_nhl = merged_nhl.drop(['team name', 'NHL', 'area', 'W', 'L'], axis = 1)
    merged_nhl = merged_nhl.groupby(['M. area', 'Population'])[['WL Ratio']].mean()
    merged_nhl = merged_nhl.reset_index()
    merged_nhl = merged_nhl.set_index('M. area')
    merged_nhl = merged_nhl.rename(columns = {'WL Ratio': 'WL Ratio NHL'})
    
    # MLB Data
    mlb = mlb_df[mlb_df['year'] == 2018]
    mlb['team name'] = mlb['team'].str.split().str[-1]
    mlb['area'] = mlb.team.str.rsplit(' ', 1).str[0]
    mlb = mlb[['team', 'W', 'L', 'team name', 'area']]
    mlb['team name'] = mlb['team name'].replace(['Sox', 'Rays', 'Jays', ''],
                                                ['Red Sox', 'Bay Rays', 'Blue Jays', ''])
    mlb['team name'] = mlb['team name'].replace([])
    mlb['team name'][8] = 'White Sox'
    mlb = mlb.set_index('team name')
    # Cities Data
    cities_mlb = cities.drop(['NHL', 'NBA', 'NFL'], axis = 1)
    cities_mlb.rename(columns={'Metropolitan area': 'M. area', 'Population (2016 est.)[8]': 'Population', 'MLB': 'team name'}, inplace = True)
    cities_mlb['team name'] = cities_mlb['team name'].str.replace(r'\[(.*?)\]', '')
    cities_mlb = (cities_mlb.set_index(['M. area', 'Population']).apply(lambda x: x.str.split('[a-z][A-Z]').explode()).reset_index())
    cities_mlb['team name'] = cities_mlb['team name'].replace(['Yankee', 'ets', 'Dodger', 'ngels', 'Giant', 'thletics', 'Cub', 'hite Sox', 'Rays'],
                                                ['Yankees', 'Mets', 'Dodgers', 'Angels', 'Giants', 'Athletics', 'Cubs', 'White Sox', 'Bay Rays'])
    cities_mlb = cities_mlb.set_index('team name')
    # Merging and WL-Ratio
    merged_mlb = pd.merge(cities_mlb, mlb, how = 'inner', left_index = True, right_index = True)
    merged_mlb = merged_mlb.reset_index()
    merged_mlb = merged_mlb.sort_values('M. area')
    merged_mlb = merged_mlb.set_index('M. area')
    merged_mlb['W'] = pd.to_numeric(merged_mlb['W'], downcast = 'float')
    merged_mlb['L'] = pd.to_numeric(merged_mlb['L'], downcast = 'float')
    merged_mlb['Population'] = pd.to_numeric(merged_mlb['Population'], downcast = 'float')
    merged_mlb['WL Ratio'] = merged_mlb['W']/(merged_mlb['W'] + merged_mlb['L'])
    merged_mlb = merged_mlb.drop(['team name', 'team', 'area', 'W', 'L'], axis = 1)
    merged_mlb = merged_mlb.groupby(['M. area', 'Population'])[['WL Ratio']].mean()
    merged_mlb = merged_mlb.reset_index()
    merged_mlb = merged_mlb.set_index('M. area')
    merged_mlb = merged_mlb.rename(columns = {'WL Ratio': 'WL Ratio MLB'})
    
    #NFL Data
    # Only data from 2018
    nfl = nfl_df[nfl_df['year'] == 2018]
    nfl = nfl[['W', 'L', 'team']]
    nfl['team'] = nfl['team'].str.replace("*", '')
    nfl['team'] = nfl['team'].str.replace('+', '')
    nfl['team name'] = nfl['team'].str.split().str[-1]
    nfl['area'] = nfl.team.str.rsplit(' ', 1).str[0]
    nfl = nfl.set_index('team name')
    #Cities Data
    cities_nfl = cities.drop(['NHL', 'NBA', 'MLB'], axis = 1)
    cities_nfl.rename(columns={'Metropolitan area': 'M. area', 'Population (2016 est.)[8]': 'Population', 'NFL': 'team name'}, inplace = True)
    cities_nfl['team name'] = cities_nfl['team name'].str.replace(r'\[(.*?)\]', '')
    cities_nfl
    cities_nfl = (cities_nfl.set_index(['M. area', 'Population']).apply(lambda x: x.str.split('[a-z][A-Z]').explode()).reset_index())
    cities_nfl
    cities_nfl['team name'] = cities_nfl['team name'].replace(['Giant', 'ets', 'Ram', 'hargers', '49er', 'aiders'],
                                                            ['Giants', 'Jets', 'Rams', 'Chargers', '49ers', 'Raiders'])
    cities_nfl = cities_nfl.set_index('team name')
    #Merging
    merged_nfl = pd.merge(cities_nfl, nfl, how = 'inner', left_index = True, right_index = True)
    merged_nfl = merged_nfl.reset_index()
    merged_nfl = merged_nfl.sort_values('M. area')
    merged_nfl = merged_nfl.set_index('M. area')
    merged_nfl['W'] = pd.to_numeric(merged_nfl['W'], downcast = 'float')
    merged_nfl['L'] = pd.to_numeric(merged_nfl['L'], downcast = 'float')
    merged_nfl['Population'] = pd.to_numeric(merged_nfl['Population'], downcast = 'float')
    merged_nfl['WL Ratio'] = merged_nfl['W']/(merged_nfl['W'] + merged_nfl['L'])
    merged_nfl = merged_nfl.drop(['team name', 'team', 'area', 'W', 'L'], axis = 1)
    merged_nfl = merged_nfl.groupby(['M. area', 'Population'])[['WL Ratio']].mean()
    merged_nfl = merged_nfl.reset_index()
    merged_nfl = merged_nfl.set_index('M. area')
    merged_nfl = merged_nfl.rename(columns = {'WL Ratio': 'WL Ratio NFL'})
    
    # Note: p_values is a full dataframe, so df.loc["NFL","NBA"] should be the same as df.loc["NBA","NFL"] and
    # df.loc["NFL","NFL"] should return np.nan
    sports = ['NFL', 'NBA', 'NHL', 'MLB']
    p_values = pd.DataFrame({k:np.nan for k in sports}, index=sports)
    
    # NBA-NHL
    merged_nba_nhl = pd.merge(merged_nba, merged_nhl, how ='inner', right_index = True, left_index = True)
    p_values['NBA']['NHL'] = stats.ttest_rel(merged_nba_nhl['WL Ratio NBA'],merged_nba_nhl['WL Ratio NHL'])[1]
    p_values['NHL']['NBA'] = p_values['NBA']['NHL']
    # MLB-NFL
    merged_mlb_nfl = pd.merge(merged_mlb, merged_nfl, how = 'inner', right_index = True, left_index = True)
    p_values['MLB']['NFL'] = stats.ttest_rel(merged_mlb_nfl['WL Ratio MLB'], merged_mlb_nfl['WL Ratio NFL'])[1]
    p_values['NFL']['MLB'] = p_values['MLB']['NFL']
    # NFL-NBA
    merged_nfl_nba = pd.merge(merged_nfl, merged_nba, how = 'inner', right_index = True, left_index = True)
    p_values['NFL']['NBA'] = stats.ttest_rel(merged_nfl_nba['WL Ratio NFL'], merged_nfl_nba['WL Ratio NBA'])[1]
    p_values['NBA']['NFL'] = p_values['NFL']['NBA']
    # NFL-NHL
    merged_nfl_nhl = pd.merge(merged_nfl, merged_nhl, how = 'inner', right_index = True, left_index = True)
    p_values['NFL']['NHL'] = stats.ttest_rel(merged_nfl_nhl['WL Ratio NFL'], merged_nfl_nhl['WL Ratio NHL'])[1]
    p_values['NHL']['NFL'] = p_values['NFL']['NHL']
    # NBA-MLB
    merged_nba_mlb = pd.merge(merged_nba, merged_mlb, how = 'inner', right_index = True, left_index = True)
    p_values['NBA']['MLB'] = stats.ttest_rel(merged_nba_mlb['WL Ratio NBA'], merged_nba_mlb['WL Ratio MLB'])[1]
    p_values['MLB']['NBA'] = p_values['NBA']['MLB']
    # NHL-MLB
    merged_nhl_mlb = pd.merge(merged_nhl, merged_mlb, how = 'inner', right_index = True, left_index = True)
    p_values['NHL']['MLB'] = stats.ttest_rel(merged_nhl_mlb['WL Ratio NHL'], merged_nhl_mlb['WL Ratio MLB'])[1]
    p_values['MLB']['NHL'] = p_values['NHL']['MLB']
    
    #Return results
    assert abs(p_values.loc["NBA", "NHL"] - 0.02) <= 1e-2, "The NBA-NHL p-value should be around 0.02"
    assert abs(p_values.loc["MLB", "NFL"] - 0.80) <= 1e-2, "The MLB-NFL p-value should be around 0.80"
    return p_values
```
